```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# パラメータ設定
n = 10   # 試行回数
p = 0.5  # 成功確率
damage = 50

# x軸の値（0 から n までの整数）
x = np.arange(0, n+1)

# 二項分布の確率質量関数 (PMF)
y = binom.pmf(x, n, p)

# プロット
x_label = [str(i * damage) for i in x]
plt.xticks(x, x_label)

plt.bar(x, y)
plt.xlabel('Damage')
plt.ylabel('Probability')

evalue = n * p
std = np.sqrt(n * p * (1 - p))
print('期待値: ', evalue * damage)
print('標準偏差: ', std * damage)
# グラフ表示
plt.show()
```

```python
index_within_1std = np.where(np.abs(x - evalue) <= std)
print('1σ内: ', np.sum(y[index_within_1std]))
damages = x[index_within_1std] * damage
print('1σ内のダメージ: ', damages)
```

```python
max_n = 11

evalues = []
stds = []
for i in range(1, max_n + 1):
    evalues.append(i * p * damage)
    stds.append(np.sqrt(i * p * (1 - p)) * damage)

plus_stds = [evalues[i] + stds[i] for i in range(max_n)]
minus_stds = [evalues[i] - stds[i] for i in range(max_n)]
max_damages = [i * damage for i in range(1, max_n + 1)]

# plt.bar(range(1, max_n + 1), evalues, label='Expected value', color='gray')


plt.bar(range(1, max_n + 1), max_damages, label='Max damage')
plt.bar(range(1, max_n + 1), plus_stds, label='Expected value + Standard deviation', linestyle='dashed')
plt.bar(range(1, max_n + 1), evalues, label='Expected value')
plt.bar(range(1, max_n + 1), minus_stds, label='Expected value - Standard deviation', linestyle='dashed', color='gray')
plt.legend()
plt.grid()
plt.xlabel('Number of energy')
plt.ylabel('Damage')
plt.show()
requeirments
```
